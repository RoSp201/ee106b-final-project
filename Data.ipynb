```python
import matplotlib.pyplot as plt
import pandas
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

import numpy as np
```

```python
hv = np.load("human_vels25.npy")
rv = np.load("robot_vels25.npy")

hv_x = hv[::3]
hv_y = hv[1::3]
hv_z = hv[2::3]

rv_x = rv[::3]
rv_y = rv[1::3]
rv_z = rv[2::3]
```

```python
plt.figure()
plt.plot([i for i in range(len(rv_x))],rv_x,color='red',label="human")
plt.plot([i for i in range(len(hv_x))],hv_x/1.6,color='black',label="Baxter")
plt.title('Human and Baxter Linear Velocity X')
plt.xlabel('Samples')
plt.ylabel('Velocity(m/s)')
plt.legend()
plt.savefig('xvel25.png')

plt.figure()
plt.plot([i for i in range(len(rv_y))],rv_y,color='green',label="human")
plt.plot([i for i in range(len(hv_y))],hv_y/1.6,color='black',label="Baxter")
plt.title('Human and Baxter Linear Velocity Y')
plt.xlabel('Samples')
plt.ylabel('Velocity(m/s)')
plt.legend()
plt.savefig('yvel25.png')

plt.figure()
plt.plot([i for i in range(len(rv_z))],rv_z,color='blue',label="human")
plt.plot([i for i in range(len(hv_z))],hv_z/1.6,color='black',label="Baxter")
plt.title('Human and Baxter Linear Velocity Z')
plt.xlabel('Samples')
plt.ylabel('Velocity(m/s)')
plt.legend()
plt.savefig('zvel25.png')
```

```python
x_diff = (hv_x-rv_x)
y_diff = (hv_y-rv_y)
z_diff = (hv_z-rv_z)
x_norm = (x_diff - np.min(x_diff))/(np.max(x_diff)- np.min(x_diff))
y_norm = (y_diff - np.min(y_diff))/(np.max(y_diff)- np.min(y_diff))
z_norm = (z_diff - np.min(z_diff))/(np.max(z_diff)- np.min(z_diff))
print(np.sqrt(np.sum((x_norm)**2))/x_norm.shape[0])
print(np.sqrt(np.sum((y_norm)**2))/y_norm.shape[0])
print(np.sqrt(np.sum((z_norm)**2))/z_norm.shape[0])
```
