```python
import matplotlib.pyplot as plt
import pandas
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

import numpy as np
```

```python
hv = np.load("human_vels25.npy")
rv = np.load("robot_vels25.npy")

hv_x = hv[::3]
hv_y = hv[1::3]
hv_z = hv[2::3]

rv_x = rv[::3]
rv_y = rv[1::3]
rv_z = rv[2::3]
```

```python
plt.figure()
plt.plot([i for i in range(len(rv_x))],rv_x,color='red')
plt.plot([i for i in range(len(hv_x))],hv_x/1.6,color='black')
plt.figure()
plt.plot([i for i in range(len(rv_y))],rv_y,color='green')
plt.plot([i for i in range(len(hv_y))],hv_y/1.6,color='black')
plt.figure()

plt.plot([i for i in range(len(rv_z))],rv_z,color='blue')
plt.plot([i for i in range(len(hv_z))],hv_z/1.6,color='black')
```

```python
print(np.sqrt(np.sum((hv_x-rv_x)**2))/hv_x.shape[0])
print(np.sqrt(np.sum((hv_y-rv_y)**2))/hv_y.shape[0])
print(np.sqrt(np.sum((hv_z-rv_z)**2))/hv_z.shape[0])
```
